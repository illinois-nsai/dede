```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("data_updated.csv").set_index(["num_cpus", "n"])
df["misc_time"] = df["full_time"] - df["solve_time"] - df["group_time"]
df = df.sort_index()
df
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

df_plot = df.unstack(level=0)

df_plot["full_time"].plot(ax=ax, marker="o")
ax.legend(df_plot.columns.get_level_values(1), title="num_cpus")

ax.set_xlabel("Size of problem")
ax.set_ylabel("Full Time (s)")

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

df_plot = df.unstack(level=0)

(df_plot["full_time"] - df_plot["group_time"]).plot(ax=ax, marker="o")
ax.legend(df_plot.columns.get_level_values(1), title="num_cpus")

ax.set_xlabel("Size of problem")
ax.set_ylabel("Non-group time (s)")

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

df_plot = df.unstack(level=0)

df_plot["group_time"].plot(ax=ax, marker="o")
ax.legend(df_plot.columns.get_level_values(1), title="num_cpus")

ax.set_xlabel("Size of problem")
ax.set_ylabel("Group Objectives Time (s)")

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

df_plot = df.unstack(level=0)

df_plot["solve_time"].plot(ax=ax, marker="o")
ax.legend(df_plot.columns.get_level_values(1), title="num_cpus")

ax.set_xlabel("Size of problem")
ax.set_ylabel("Solve Time (s)")

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

df_plot = df.unstack(level=0)

(df_plot["misc_time"]).plot(ax=ax, marker="o")
ax.legend(df_plot.columns.get_level_values(1), title="num_cpus")

ax.set_xlabel("Size of problem")
ax.set_ylabel("Misc Time (s)")

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

df_plot = df.unstack(level=0)

(df_plot["solve_time"] / df_plot["full_time"]).plot(ax=ax, marker="o")
ax.legend(df_plot.columns.get_level_values(1), title="num_cpus")

ax.set_xlabel("Size of problem")
ax.set_ylabel("Fraction of time used to solve")

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

df_plot = df.unstack(level=0)

(df_plot["group_time"] / df_plot["full_time"]).plot(ax=ax, marker="o")
ax.legend(df_plot.columns.get_level_values(1), title="num_cpus")

ax.set_xlabel("Size of problem")
ax.set_ylabel("Fraction of time used to group objectives")

fig.show()
```
